```python
import pandas as pd
import matplotlib.pyplot as plt
a = {'values': [18, 21, 22, 25, 28, 30, 32, 34, 35, 36, 38, 40, 42, 45, 50, 55, 60, 65, 70]}
a1 = pd.DataFrame(a)

# plot a histogram
plt.hist(a1['values'], bins=10)
plt.show()
```

```python
'''customized histogram'''
b = {'values': [23, 45, 30, 50, 67, 35, 47, 62, 25, 58, 42, 36, 53, 68, 32]}
b1 = pd.DataFrame(b)

# plot a customized histogram
plt.hist(b1['values'], bins=6 , edgecolor='black', color='blue', alpha=0.7)
plt.title('Customized Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
