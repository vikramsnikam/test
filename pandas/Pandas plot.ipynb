```python
import pandas as pd
import matplotlib.pyplot as plt
car = ["Caterham", "Tesla", "Audi", "BMW", "Ford", "Jeep"]
weight = [0.48, 1.7, 2, 2, 2.3, 3]

# create a DataFrame
a = {'Car': car, 'Weight': weight}
a1 = pd.DataFrame(a)
'''line chart'''
# plot using Pandas
a1.plot(x='Car', y='Weight', kind='line', marker='o')
plt.xlabel('Car')
plt.ylabel('Weight')
plt.title('Car Weights')
plt.show()
```

```python
'''scatter plot'''
a1.plot(x='Car', y='Weight', kind='scatter', marker='o', color='red')
plt.xlabel('Car')
plt.ylabel('Weight')
plt.title('Car Weights (Scatter Plot)')
plt.grid(True)
plt.show()
```

```python
'''Bar graph'''
a1.plot(x='Car', y='Weight', kind='bar', color='green')
plt.xlabel('Car')
plt.ylabel('Weight')
plt.title('Car Weights (Bar Graph)')
plt.tight_layout()
plt.show()
```
