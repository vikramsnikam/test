```python
'''Measures of Central Tendency'''
'''mean and median'''
import numpy as np
marks = [10, 20, 30, 40, 50]
print("Mean =",np.mean(marks))
print("Median =",np.median(marks))
'''mode '''
from statistics import mode
Example=[2, 2, 3, 3, 3, 4]
print("Mode =", mode([2,2,3,3,3,4]))
```

```python
'''Measure of dispersion'''
a=[10,20,30,40,50]
'''range'''
print('range=',max(a)-min(a))
'''variance'''
print()
print('''statistics.stdev(a)
 computes the sample standard deviation
numpy.std(a)
 by default computes the population standard deviation ''')
print()
print('variance',np.var(a))
import statistics as st
print('sample variance',st.variance(a))
'''calculation'''
# a = [10, 20, 30, 40, 50]

#Mean: x̄ = (10 + 20 + 30 + 40 + 50) / 5 = 30

#Squared deviations sum: (20² + 10² + 0² + 10² + 20²) = 1000

#Population variance: σ² = 1000 / 5 = 200

#Sample variance: s² = 1000 / (5 - 1) = 250
'''std deviation'''
'''square root of variance'''
print('standard sample deviation',st.stdev(a))
print("Standard Deviation =", np.std(a))

```

```python
'''IQR'''
Q1 = np.percentile(marks, 25)
print(Q1)
Q3 = np.percentile(marks, 75)
print(Q3)
IQR = Q3 - Q1
print("IQR =", IQR)
'''For a = [10, 20, 30, 40, 50]:
Q1 = 20
Q3 = 40
IQR = 20'''
```

```python
'''Skewness'''
from scipy.stats import skew
marks = [55, 60, 62, 65, 70, 72, 90, 10]
print("Skewness =", skew(marks))
import matplotlib.pyplot as plt
```

```python
'''Kurtosis'''
'''high kurtosis mean sharp peak data centered around mean'''
'''low kurtosis mean falt  data spread data'''
# Most values near 85, with few extreme outliers
data_high = [85, 86, 84, 87, 85, 0, 100]
plt.hist(data_high, bins=10, color="skyblue", edgecolor="black")
plt.title("High Kurtosis (Leptokurtic)")
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.show()
# Spread out values from 0 to 100
data_low = np.linspace(0, 10, 200) # 200 evenly spaced points between 0 and 100
plt.hist(data_low, bins=20, color="salmon", edgecolor="black")
plt.title("Low Kurtosis (Platykurtic)")
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.show()
```
