```python
'''Spatial data refers to data that is represented in a geometric space.'''
import numpy as np
from scipy.spatial import Delaunay
import matplotlib.pyplot as plt

a = np.array([
  [2, 4],
  [3, 4],
  [3, 0],
  [2, 2],
  [4, 1],

])

a1 = Delaunay(a).simplices

plt.triplot(a[:, 0], a[:, 1], a1)
plt.scatter(a[:, 0], a[:, 1], color='r')

plt.show()
```

```python
'''Convex Hull
A convex hull is the smallest polygon that covers all of the given points.'''
from scipy.spatial import ConvexHull
b = np.array([
  [2, 4],
  [3, 4],
  [3, 0],
  [2, 2],
  [4, 1],
  [1, 2],
  [5, 0],
  [3, 1],
  [1, 2],
  [0, 2]
])

b1 = ConvexHull(b)
hull_points = b1.simplices

plt.scatter(b[:,0], b[:,1])
for simplex in hull_points:
  plt.plot(b[simplex,0], b[simplex,1], 'k-')

plt.show()
```

```python
'''KD tree'''
from scipy.spatial import KDTree

# Define points in 2D
c = np.array([
    [2, 3],
    [5, 4],
    [9, 6],
    [4, 7],
    [8, 1],
    [7, 2]
])
print(c)
# Build KDTree
tree = KDTree(c)

# Query: nearest neighbor of (9,2)
dist, idx = tree.query([9, 2])
print("Nearest neighbor:", c[idx], "with distance:", dist)
```

```python
# Find all points within radius=3 of (9,2)
idxs = tree.query_ball_point([9, 2], r=3)
print("Neighbors within radius=3:", c[idxs])
```
