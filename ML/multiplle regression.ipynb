```python
'''Multiple Linear Regression '''
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, mean_squared_error
from sklearn.metrics import r2_score
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Sample data
data = {
    'study_hours': [2, 3, 4, 5, 6, 7, 8],
    'sleep_hours': [8, 7, 6, 6, 5, 5, 4],
    'marks': [50, 55, 60, 65, 70, 75, 85]
}
df = pd.DataFrame(data)

X = df[['study_hours', 'sleep_hours']]
y = df['marks']

# Scatter plot before fit
plt.scatter(df['study_hours'], y, color='blue', label='Study vs Marks')
plt.scatter(df['sleep_hours'], y, color='red', label='Sleep vs Marks')
plt.xlabel('Hours')
plt.ylabel('Marks')
plt.title('Raw Data for Multiple Linear Regression')
plt.legend()
plt.show()

```

```python
# Fit model
mlr = LinearRegression()
mlr.fit(X, y)
```

```python
print('Coefficients:', mlr.coef_)
print('Intercept:', mlr.intercept_)
print('Prediction for (study=6, sleep=5):', mlr.predict([[6, 5]]))
```

```python
y_pred = mlr.predict(X)
plt.scatter(df['study_hours'], y, color='blue', label='Actual (Study vs Marks)')
plt.plot(df['study_hours'], y_pred, color='blue', linestyle='dashed', label='Predicted vs Study')

plt.scatter(df['sleep_hours'], y, color='red', label='Actual (Sleep vs Marks)')
plt.plot(df['sleep_hours'], y_pred, color='red', linestyle='dashed', label='Predicted vs Sleep')

plt.xlabel('Hours')
plt.ylabel('Marks')
plt.title('Multiple Linear Regression (2D Projection)')
plt.legend()
plt.show()
```
