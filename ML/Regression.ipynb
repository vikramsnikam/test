```python
'''Linear regression '''
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
data = {
    'mileage': [15000, 30000, 45000, 60000, 75000],
    'price': [550000, 480000, 430000, 390000, 350000]
}

df = pd.DataFrame(data)
X = df[['mileage']]
y = df['price']


plt.scatter(df.mileage, df.price, color='red', marker='+')
```

```python
X = df[['mileage']]
y = df['price']

reg = LinearRegression()
reg.fit(X, y)
```

```python
a=reg.predict([[45000]])
print(a)
b=reg.predict([[100000]])
print(b)
a1=reg.coef_
b1=reg.intercept_
print(a1)
print(b1)
print("Coefficients:", reg.coef_)
```

```python

reg = LinearRegression()
reg.fit(X, y)
# Predict values
y_pred = reg.predict(X)

# Plot scatter points
plt.scatter(X, y, color='blue', label='Actual data')

# Plot regression line
plt.plot(X, y_pred, color='red', linewidth=2, label='Regression line')

plt.xlabel('Mileage')
plt.ylabel('Price')
plt.title('Car Price vs Mileage with Regression Line')
plt.legend()
plt.show()
```
